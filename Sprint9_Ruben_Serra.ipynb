1. Calculadora de l'√≠ndex de massa corporal

Escriu una funci√≥ que calculi l'IMC ingressat per l'usuari/√†ria, √©s a dir, qui ho executi haur√† d'ingressar aquestes dades. Pots obtenir m√©s informaci√≥ del seu c√†lcul en: üîó √çndice de masa corporal IMC que es y como se calcula.

La funci√≥ ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades pr√®viament, perqu√® envi√Ø un missatge d‚Äôadvert√®ncia si les dades introdu√Ødes per l‚Äôusuari/a no estan en el format adequat o no pren valors raonables.

```python
def calcular_imc():
    
    dato_altura = input("Introduce tu altura en metros (ej. 1.75 o 1,75): ")
    dato_altura = dato_altura.replace(",", ".")

    try:
        altura = float(dato_altura)
        if altura <= 0 or altura > 2.6:
            raise ValueError
    except ValueError:
        print("ERROR: La altura debe estar en formato num√©rico y entre 0 y 2.6 metros.")
        return 

    dato_peso = input("Introduce tu peso en kg: ")

    try:
        peso = float(dato_peso)
        if peso <= 0 or peso > 600:
            raise ValueError
    except ValueError:
        print("ERROR: El peso debe ser un n√∫mero v√°lido entre 0 y 600 kg.")
        return

    imc = peso / (altura ** 2)

    if imc <18.5:
        print("Bajo peso")
    elif 18.5 <= imc <= 24.9:
        print(f"Tu IMC es: {imc:.2f} est√°s saludable")
    elif 25 <= imc <= 29.9:
        print(f"Tu IMC es: {imc:.2f} est√°s en sobrepeso")
    else:
        print(f"Tu IMC es: {imc:.2f} est√°s en Obesidad")
    

calcular_imc()
```

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les m√©s comunes s√≥n Celsius (¬∞C), Fahrenheit (¬∞F) i Kelvin (K). Tamb√© existeixen altres unitats com Rankine (¬∞Ra) i R√©aumur (¬∞Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a m√≠nim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades pr√®viament, perqu√® envi√Ø un missatge d‚Äôadvert√®ncia si les dades introduides per l‚Äôusuari/a no estan en el format adequat.

```python
def convertir_temperatura1():

    while True:
        unidad = input("Introduce la unidad de la temperatura (C, F, K, RA): ").strip().upper()
        if unidad in ("C", "F", "K", "RA"):
            break
        print("ERROR: Unidad no v√°lida. Usa C, F, K o RA.")

    while True:
        try:
            valor = float(input(f"Introduce la temperatura en {unidad}: "))
        except ValueError:
            print("ERROR: Debes introducir un n√∫mero v√°lido.")
            continue

        if unidad == "C" and not (-273.15 <= valor <= 100):
            print("ERROR: En Celsius tiene que estar entre -273.15¬∞C y 100¬∞C.")
            continue
        if unidad == "F" and not (-459.67 <= valor <= 212):
            print("ERROR: En Fahrenheit debe estar entre -459.67¬∞F y 212¬∞F.")
            continue
        if unidad == "K" and not (0 <= valor <= 373.15):
            print("ERROR: En Kelvin debe estar entre 0 K y 373.15 K.")
            continue
        if unidad == "RA" and not (0 <= valor <= 671.67):
            print("ERROR: En Rankine debe estar entre 0 Ra y 671.67 Ra.")
            continue
        break 

    match unidad:
        case "C":
            c = valor
            f = c * 9/5 + 32
            k = c + 273.15
            ra = (c + 273.15) * 9/5

        case "F":
            f = valor
            c = (f - 32) * 5/9
            k = c + 273.15
            ra = f + 459.67

        case "K":
            k = valor
            c = k - 273.15
            f = c * 9/5 + 32
            ra = k * 9/5

        case "RA":
            ra = valor
            k = ra * 5/9
            c = k - 273.15
            f = ra - 459.67

   
    conversiones = {"Temperaturas": {"Celsius": round(c, 2),"Fahrenheit": round(f, 2),
                                     "Kelvin": round(k, 2),"Rankine": round(ra, 2)}}

    return conversiones
convertir_temperatura1()
```

3. Comptador de paraules d'un text.

Escriu una funci√≥ que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casu√≠stiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina √©s la llargada mitjana de les paraules del text que has escrit? ‚ÄúHola com va?‚Äù hauria de retornar (4+3+2) / 3 = 3

```python

frase = input("Introduce una frase")

puntuacion = "?!,.:;¬°¬ø()[]{}\"'"

frase_limpia = frase

for signo in puntuacion:
    frase_limpia = frase_limpia.replace(signo, "") 

palabras = frase_limpia.split() 

contador_palabras = {}
contador_letras = {}

total_letras = 0

for palabra in palabras:
    
    if palabra in contador_palabras:
        contador_palabras[palabra] += 1 
    else:
        contador_palabras[palabra] = 1
    num_letras = len(palabra) 
    contador_letras[palabra] = num_letras

    total_letras += num_letras

promedio = total_letras / len(palabras)

print(f"Conteo de palabras: {contador_palabras}")
print(f"Letras por palabra: {contador_letras}")
print(f"Promedio de letras por palabra: {promedio:.2f}")
```

2. Conversi√≥ de tipus de dades.
El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

```python
lista = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

flotantes = []

def extraer_flotantes(obj):
    
    if isinstance(obj, (float, int)):
        flotantes.append(float(obj))
    elif isinstance(obj, str):
        try:
            f = float(obj)
            flotantes.append(f)
        except ValueError:
            pass
    elif isinstance(obj, (list, tuple)):
        for item in obj:
            extraer_flotantes(item)

extraer_flotantes(lista)

print(flotantes)
```

1. Generador de Contrasenyes
En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d‚Äôaniversari o similars. Per a solucionar-ho ens ha encarregat una funci√≥ de Python que generi contrasenyes m√©s segures. La funci√≥ ha de dependre dels seg√ºents par√†metres:
longitud (int): 
‚Ä¢ Longitud de la contrasenya
‚Ä¢ majuscules (bool = True): Si hi ha d‚Äôapar√®ixer maj√∫scules
‚Ä¢ minuscules (bool = True): Si hi ha d‚Äôapar√®ixer min√∫scules
‚Ä¢ numeros (bool = True): Si hi ha d‚Äôapar√®ixer n√∫meros
‚Ä¢ signes (bool = False): Si hi ha d‚Äôapar√®ixer car√†cters especials (,-$? o similars)

```python
import random

MAJUSCULES = [chr(i) for i in range(ord('A'), ord('Z') + 1)]
MINUSCULES = [chr(i) for i in range(ord('a'), ord('z') + 1)]
NUMEROS = [str(i) for i in range(10)]
SIGNES = list("!@#$%^&*()-_=+,.?/")

def crear_contrasenya(longitud, majuscules, minuscules, numeros, signes):

    grups = []
    if majuscules:
        grups.append(MAJUSCULES)
    if minuscules:
        grups.append(MINUSCULES)
    if numeros:
        grups.append(NUMEROS)
    if signes: 
        grups.append(SIGNES)

    if not grups:
        raise ValueError("S'ha d'activar almenys un tipus de car√†cter.")

    contrasenya = []

    for grup in grups:
        contrasenya.append(random.choice(grup))

    pool = []
    for grup in grups:
        pool.extend(grup)

    while len(contrasenya) < longitud:
        contrasenya.append(random.choice(pool))

    random.shuffle(contrasenya)

    return "".join(contrasenya)
crear_contrasenya(10, True, True, True, True)


```

La classificaci√≥ global (cada vict√≤ria: 3 pts, empat 1 pts, derrota 0 pts)

```python
import pandas as pd
ruta = r"C:\Users\ruben\Downloads\historic partits (2).txt"
archivo = pd.read_csv(ruta,
                sep="\t",
                header=None,
                names=['Equip local','Resultat','Equip visitant'],
                encoding="utf-8")

def clasi(archivo):
    archivo[["Goles Local", "Goles Visitante"]] = archivo["Resultat"].str.split("-", expand=True).astype(int)
    archivo["puntos_local"] = 0
    archivo.loc[archivo["Goles Local"] > archivo["Goles Visitante"], "puntos_local"] = 3
    archivo.loc[archivo["Goles Local"] == archivo["Goles Visitante"], "puntos_local"] = 1
    archivo["puntos_visitante"] = 0
    archivo.loc[archivo["Goles Visitante"] > archivo["Goles Local"], "puntos_visitante"] = 3
    archivo.loc[archivo["Goles Visitante"] == archivo["Goles Local"], "puntos_visitante"] = 1
    Total_puntos_visitantes = archivo.groupby("Equip visitant")["puntos_visitante"].sum()
    Total_puntos_locales = archivo.groupby("Equip local")["puntos_local"].sum()
    Clasificacion = Total_puntos_locales.add(Total_puntos_visitantes, fill_value=0).to_frame(name="Puntos")
    Goles_encajados_local = archivo.groupby("Equip local")["Goles Visitante"].sum()
    Goles_encajados_visitante = archivo.groupby("Equip visitant")["Goles Local"].sum()
    Total_goles_encajados = Goles_encajados_local.add(Goles_encajados_visitante, fill_value=0).to_frame(name="Goles en contra")
    Total_goles_encajados
    Goles_marcados_local = archivo.groupby("Equip local")["Goles Local"].sum()
    Goles_marcados_visitante = archivo.groupby("Equip visitant")["Goles Visitante"].sum()
    Total_goles_marcados = Goles_marcados_local.add(Goles_marcados_visitante, fill_value=0).to_frame(name="Goles a favor")
    Clasificacion_completa1 = pd.concat([Clasificacion, Total_goles_marcados, Total_goles_encajados], axis=1).sort_values(by= "Puntos", ascending= False)
    print (Clasificacion_completa1)
    Equipo_mas_goleador = Total_goles_marcados["Goles a favor"].idxmax()
    print(f'El equipo que m√°s goles ha metido es: {Equipo_mas_goleador}')
    Equipo_mas_goles_encajados = Total_goles_encajados["Goles en contra"].idxmax()
    print(f'El equipo que m√°s goles recibe es: {Equipo_mas_goles_encajados}')
clasi(archivo)
```
